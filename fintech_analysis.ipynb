```python
import pandas as pd

def get_data():
    investment_df = pd.read_csv('fintech_investment.csv')
    
    investment_df['blockchain_as_pct'] = (investment_df.blockchain_investment / investment_df.total_investment) * 100

    return investment_df.head()

get_data()
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import numpy as np

def build_chart():
    df = get_data()
    
    x = np.arange(len(df.year))
    width = 0.2
    
    fig, ax = plt.subplots()
    blockchain_investment = ax.bar(x - width / 2, df.blockchain_investment, width, label="Blockchain Investment $B")
    total_investment = ax.bar(x, df.total_investment, width, label="Total FinTech Investment $B")
    
    ax.set_ylabel('Total Global Investment $B')
    ax.set_title('Total Global Investment & Total Blockchain Investment $B')
    ax.set_xticks(x)
    ax.set_xticklabels(df.year)
    ax.legend()
    
    ax.bar_label(blockchain_investment, label_type='edge')
    ax.bar_label(total_investment, label_type='center')
    
    fig.set_size_inches(18.5, 10.5)
    plt.show()
    
build_chart()
```
